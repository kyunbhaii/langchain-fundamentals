```python
import random

class NakliLLM():

    def __init__(self):
        print("LLM Created")

    def predict(self, prompt):

        response_list = [
            'Delhi is the capital of India',
            'IPL is a Cricket League',
            'AI stands for Artificial Intelligence'
        ]

        return {'response':random.choice(response_list)}
```

```python
llm = NakliLLM()
```

```python
llm.predict('AI stands for?')
```

```python
class NakliPromptTemplate():

    def __init__(self, template, input_variable):
        self.template = template
        self.input_variable = input_variable
    
    def format(self, input_dict):
        return self.template.format(**input_dict)
```

```python
template = NakliPromptTemplate(
    template='Write a poem aobut {topic}',
    input_variable=['topic']
)
```

```python
prompt = template.format({'topic':'India'})
```

```python
model = NakliLLM()
```

```python
model.predict(prompt)
```

```python
class NakliLLMChain:

    def __init__(self, llm, prompt):
        self.llm = llm
        self.prompt = prompt

    def run(self, input_dict):

        final_prompt = self.prompt.format(input_dict)
        result = self.llm.predict(final_prompt)

        return result['response']
```

```python
template = NakliPromptTemplate(
    template='Write a poem aobut {topic}',
    input_variable=['topic']
)
```

```python
llm = NakliLLM()
```

```python
chain = NakliLLMChain(llm, template)
```

```python
chain.run({'topic':'AI'})
```

#### Runnables

```python
from abc import ABC, abstractmethod

class Runnables(ABC):

    @abstractmethod
    def invoke(input_data):
        pass
```

```python
class NakliLLM(Runnables):

    def __init__(self):
        print("LLM Created")

    def invoke(self, prompt):

        response_list = [
            'Delhi is the capital of India',
            'IPL is a Cricket League',
            'AI stands for Artificial Intelligence'
        ]

        return {'response':random.choice(response_list)}

    def predict(self, prompt):

        response_list = [
            'Delhi is the capital of India',
            'IPL is a Cricket League',
            'AI stands for Artificial Intelligence'
        ]

        return {'response':random.choice(response_list)}
```

```python
class NakliPromptTemplate(Runnables):

    def __init__(self, template, input_variable):
        self.template = template
        self.input_variable = input_variable
    
    def invoke(self, input_dict):
        return self.template.format(**input_dict)
    
    def format(self, input_dict):
        return self.template.format(**input_dict)
```

```python
class NakliStrOutputParser(Runnables):

    def __init__(self):
        pass

    def invoke(self, input_data):
        return input_data['response']
```

```python
class RunnableConnector(Runnables):

    def __init__(self, runnable_list):
        self.runnable_list = runnable_list
    
    def invoke(self, input_data):
        for runnable in self.runnable_list:
            input_data = runnable.invoke(input_data)

        return input_data
```

```python
prompt = NakliPromptTemplate(
    template='Write something about {topic}',
    input_variable=['topic']
)
```

```python
llm = NakliLLM()
```

```python
parser = NakliStrOutputParser()
```

```python
chain = RunnableConnector([prompt, llm, parser])
```

```python
chain.invoke({'topic':'AI'})
```
