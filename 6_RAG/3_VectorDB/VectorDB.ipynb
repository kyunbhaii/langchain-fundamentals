```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_chroma import Chroma
```

```python
embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
from langchain_core.documents import Document

# Create LangChain documents for IPL players

doc1 = Document(
    page_content="Virat Kohli is one of the most successful and consistent batsmen in IPL history. Known for his aggressive batting style and match-winning performances.",
    metadata={"team": "Royal Challengers Bangalore"}
)

doc2 = Document(
    page_content="Rohit Sharma is the most successful captain in IPL history, leading Mumbai Indians to five titles. He's known for his calm leadership and explosive batting.",
    metadata={"team": "Mumbai Indians"}
)

doc3 = Document(
    page_content="MS Dhoni, famously known as Captain Cool, has led Chennai Super Kings to multiple IPL titles. His finishing skills and wicket-keeping abilities are exceptional.",
    metadata={"team": "Chennai Super Kings"}
)

doc4 = Document(
    page_content="Jasprit Bumrah is considered one of the best fast bowlers in T20 cricket. Playing for Mumbai Indians, he is known for his yorkers and death bowling skills.",
    metadata={"team": "Mumbai Indians"}
)

doc5 = Document(
    page_content="Ravindra Jadeja is a dynamic all-rounder who contributes with both bat and ball. Representing Chennai Super Kings, his fielding skills are outstanding.",
    metadata={"team": "Chennai Super Kings"}
)
```

```python
docs = [doc1, doc2, doc3, doc4, doc5]
```

```python
vector_store = Chroma(
    embedding_function=embedding_model,
    persist_directory='chroma_db',
    collection_name='sample'
)
```

```python
# add documents
vector_store.add_documents(docs)
```

```python
# View-documents
vector_store.get(include=['embeddings', 'documents', 'metadatas'])
```

```python
# search documents
vector_store.similarity_search(
    query = "Who among these are a bowler",
    k=2
)
```

```python
# search with similarity score

vector_store.similarity_search_with_score(
    query = "Who among these are a bowler",
    k=2
)
```

```python
# meta data filtering

vector_store.similarity_search_with_score(
    query = "",
    filter={'team':'Chennai Super Kings'}
)
```

```python
# update documents

# update documents
updated_doc1 = Document(
    page_content="Virat Kohli, the former captain of Royal Challengers Bangalore (RCB), is renowned for his aggressive leadership and consistent batting performances in the IPL.",
    metadata={"team": "Royal Challengers Bangalore"}
)

vector_store.update_document(document_id='347161d2-53c0-4313-9982-3875ef208975', document=updated_doc1)
```
